```python
import pandas as pd
import matplotlib.pyplot as plt


```

```python
data = pd.read_excel("jubilee_controller-main/jubilee_controller/analise.xlsx")
```

```python
x = list(data.index)
y = data['Temperatura (°C)']

y_upper = y + 1
y_lower = y - 1

plt.figure(figsize=(10, 5))
plt.plot(x, y, color='blue')
plt.fill_between(x, y_lower, y_upper, color='blue', alpha=0.2)
plt.xlabel('Índice')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
